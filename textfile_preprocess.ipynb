```python
import json
import pandas as pd 
```

```python
#We are reading files with this code
def readFile(fileName):
        fileObj = open(fileName, "r") #opens the file in read mode
        words = fileObj.read().splitlines() #puts the file into an array
        fileObj.close()
        return words
```

```python
#Read the stored files in the array
text_file = open("5637176_1840882096_allevents.txt", "r")
lines = text_file.readlines()
print (lines[0])
print (len(lines))
text_file.close()
```

```python
#We try to find the event, restart and other data on the whole dataset of files
df = pd.DataFrame([{'Event': 'Restart','Time': 8.502,'AdjustCount': 5093,'Restarts': 73,'Conflicts': 20652,'Decisions': 75009,'Propagations': 38723637,'Vars': 30171,'Clauses': 93919,'Literals': 225249,'LEARNT_Limit': 98320,'LEARNT_Clauses': 19180,'LEARNT_Lit/Cl': 28,'Progress': 7.861}])
for i in range(len(lines)):
    str = ''
    x = (lines[i].strip('[\n').split(',')[0:-1])
    for i in range(len(x)):
        str += x[i]+','
    str = str[:-1]
    try:
        y = json.loads(str)
    except:
        continue

    df1 = pd.DataFrame([y])
    df = pd.concat([df, df1], ignore_index = True, axis = 0)


df
```
