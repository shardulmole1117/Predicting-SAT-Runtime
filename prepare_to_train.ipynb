```python
import pandas as pd 
import numpy as np 
import os
```

```python
#Reading the CSV file
df = pd.read_csv('data_processed/s77-100.csv')
df
```

```python
#We take mean, median and standard deviation of the data
data = {
    'Time':df['Time'].max(),
    'AdjustCount':df['AdjustCount'].mean(),
    'AdjustCount_std':df['AdjustCount'].median(),
    'Restarts': df['Restarts'].mean(),
    'Restarts_std':df['Restarts'].std(),
    'Conflicts':df['Conflicts'].mean(),
    'Conflicts_std':df['Conflicts'].std(),
    'Decisions':df['Decisions'].mean(),
    'Decisions_std':df['Decisions'].std(),
    'Propagations':df['Propagations'].mean(),
    'Propagations_std':df['Propagations'].std(),
    'Vars':df['Vars'].mean(),
    'Clauses':df['Clauses'].mean(),
    'Literals':df['Literals'].mean(),
    'LEARNT_Limit':df['LEARNT_Limit'].mean(),
    'LEARNT_Limit_std':df['LEARNT_Limit'].std(),
    'LEARNT_Clauses':df['LEARNT_Clauses'].mean(),
    'LEARNT_Clauses_std':df['LEARNT_Clauses'].std(),
    'LEARNT_Lit/Cl':df['LEARNT_Lit/Cl'].mean(),
    'LEARNT_Lit/Cl_std':df['LEARNT_Lit/Cl'].std(),
    'Progress':df['Progress'].mean(),
    'timecpu':df['timecpu'].mean()


}
df1 = pd.DataFrame([data])
df1
```

```python
#Processing all the files in the folder
files = os.listdir('data_processed')
for i in range(1,len(files)):
    df = pd.read_csv(f'data_processed/{files[i]}')
    data = {
    'Time':df['Time'].max(),
    'AdjustCount':df['AdjustCount'].mean(),
    'AdjustCount_std':df['AdjustCount'].median(),
    'Restarts': df['Restarts'].mean(),
    'Restarts_std':df['Restarts'].std(),
    'Conflicts':df['Conflicts'].mean(),
    'Conflicts_std':df['Conflicts'].std(),
    'Decisions':df['Decisions'].mean(),
    'Decisions_std':df['Decisions'].std(),
    'Propagations':df['Propagations'].mean(),
    'Propagations_std':df['Propagations'].std(),
    'Vars':df['Vars'].mean(),
    'Clauses':df['Clauses'].mean(),
    'Literals':df['Literals'].mean(),
    'LEARNT_Limit':df['LEARNT_Limit'].mean(),
    'LEARNT_Limit_std':df['LEARNT_Limit'].std(),
    'LEARNT_Clauses':df['LEARNT_Clauses'].mean(),
    'LEARNT_Clauses_std':df['LEARNT_Clauses'].std(),
    'LEARNT_Lit/Cl':df['LEARNT_Lit/Cl'].mean(),
    'LEARNT_Lit/Cl_std':df['LEARNT_Lit/Cl'].std(),
    'Progress':df['Progress'].mean(),
    'timecpu':df['timecpu'].mean()
    }
    df2 = pd.DataFrame([data])
    df1 = pd.concat([df1, df2], ignore_index = True, axis = 0)
```
