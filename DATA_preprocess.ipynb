```python
#Creating the directory to save processed CSVs
os.mkdir('data_processed')
```

```python
#Reading each CSV and then writing the processed CSV to the data_processed directory
for i in range(len(csvs)):
    df = pd.read_csv(os.path.normpath(os.path.join('DATA',csvs[i])), sep='delimiter', header=None)
    filename = csvs[i]
    time_str = df.at[df.shape[0]-2,0]
    time = float(re.findall(r"[-+]?\d*\.\d+|\d+", time_str)[0])
    df = df[11:-9]
    data = df[0].tolist()
    try:
        res = json.loads(df.at[11,0])
        df1 = pd.DataFrame([res])
        for i in range(1,len(df)):
            df2 = pd.DataFrame([json.loads(data[i])])
            df1 = pd.concat([df1, df2], ignore_index = True, axis = 0)
        df1['timecpu'] = np.array([time]*len(df))
        df1.to_csv(f'data_processed/{filename}',index=False)
    except:
        continue



    
```
